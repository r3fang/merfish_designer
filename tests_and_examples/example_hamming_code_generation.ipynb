```python
# Generate a modified Hamming distance 4 code set using the stochastic method
from MERFISH_probe_design.hamming_codes.generate_modified_hamming_code_stochastic import generate_modified_hamming_codes

code_length = 22
code_list = generate_modified_hamming_codes(code_length, 4, 4, n_threads=8)
```

```python
# Plot the distribution of pairwise Hamming distances
import MERFISH_probe_design.hamming_codes.hamming_code_analysis as hca

hca.plot_pairwise_hamming_distance_distribution(code_list)
```

```python
# Plot the coverage of each bit
hca.plot_bit_coverage(code_list, code_length)
```

```python
# Save the codebook to a JSON file
import os
import json

os.makedirs('temporary_data', exist_ok=True)

code_list = [sorted(list(c)) for c in code_list] # Convert the sets to lists for JSON serialization
with open(f'temporary_data/HM4_codes_{code_length}.json', 'w') as f:
    json.dump(code_list, f)
```
